```python
from matplotlib import pyplot as plt

def config_plot(nrow):
    plt.figure(figsize=(20,4*nrow),dpi=100)
    
def visualize(sketch,show=True):
    for xs,ys in sketch['drawing']:
        ys = [-y for y in ys]
        plt.plot(xs,ys)
    if show:
        plt.show()
```

```python
def visualize_many(data):
    n = len(data)
    ncol = 5
    nrow = n//ncol + (n%ncol > 0)
    config_plot(nrow)
    for i in range(n):
        plt.subplot(nrow,ncol,i+1)
        visualize(data[i],show=False)
    plt.show()
```

```python
visualize_many(data[:50])
```

```python
import os

base_dir = 'train_simplified'
target_dir = 'train_sampled'
valid_dir = 'valid_sampled'
test_dir = 'test_sampled'
*filelist, = os.listdir(base_dir)[1:101]
```

```python
for file in filelist:
    pd.read_csv(os.path.join(base_dir,file),nrows=100).to_csv(os.path.join(target_dir,file),index=False)
```

```python
for file in filelist:
    pd.read_csv(os.path.join(base_dir,file),nrows=10,skiprows=110).to_csv(os.path.join(test_dir,file),index=False)
```

```python
for file in filelist:
    pd.read_csv(os.path.join(base_dir,file),nrows=20,skiprows=150).to_csv(os.path.join(valid_dir,file),index=False)
```
